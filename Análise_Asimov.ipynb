```python
#Caso não queira remover os valores nulos, mas marcar com alguma identificação
df_viagens['Cargo'] = df_viagens['Cargo'].fillna('NÃO IDENTIFICADO')
df_viagens
```

```python
#Cast para data
df_viagens['Período - Data de início'] = pd.to_datetime(df_viagens['Período - Data de início'], format="%d/%m/%Y")
df_viagens['Período - Data de fim'] = pd.to_datetime(df_viagens['Período - Data de fim'], format="%d/%m/%Y")
```

```python
df_viagens
```

```python
df_viagens['Mês da viagem'] = df_viagens['Período - Data de início'].dt.month_name()
```

```python
#df_viagens['Período - Data de fim'] - df_viagens['Período - Data de início']

#Para mostrar somente os dias de viagens e ocultar os outras informações do timedelta
df_viagens['Dias da viagem'] = (df_viagens['Período - Data de fim'] - df_viagens['Período - Data de início']).dt.days
```

```python
df_viagens
```

```python
#df_viagens.drop('Dias da viagem', axis=1, inplace=True)
#excluindo coluna
```

```python
#Agregando informações(.agg)
(
    df_viagens
    .groupby('Cargo')
    .agg(
        despesa_media = ('Despesas' , 'mean'),
        duracao_media = ('Dias da viagem', 'mean'),
        despesas_totais = ('Despesas', 'sum'),
        destino_mais_frequente = ('Destinos', pd.Series.mode), #pd.Series.mode para calcular moda
        n_viagens = ('Nome', 'count')
        )
    .reset_index()
        )
```

```python

df_cargos = df_viagens['Cargo'].value_counts(normalize=True).reset_index()
df_cargos
```

```python
df_cargos.loc[df_cargos['proportion']>0.01]
```

```python
df_cargos.loc[df_cargos['proportion']>0.01, 'Cargo']
```

```python
df_viagens_consolidado = (
    df_viagens
    .groupby('Cargo')
    .agg(
        despesa_media = ('Despesas' , 'mean'),
        duracao_media = ('Dias da viagem', 'mean'),
        despesas_totais = ('Despesas', 'sum'),
        destino_mais_frequente = ('Destinos', pd.Series.mode), #pd.Series.mode para calcular moda
        n_viagens = ('Nome', 'count')
        )
    .reset_index()
        )
```

```python
cargos_relevantes = df_cargos.loc[df_cargos['proportion']>0.01, 'Cargo']
```

```python
filtro = df_viagens_consolidado['Cargo'].isin(cargos_relevantes)
```

```python
df_final = df_viagens_consolidado[filtro]
```

```python
df_final
```

```python
df_final= df_final.sort_values(by = 'n_viagens', ascending=False)
```

```python
df_final.plot(x = 'Cargo', y = 'n_viagens', kind = 'bar')
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(16,6))
ax.barh(df_final['Cargo'], df_final['n_viagens'], color='#1f5c23')
ax.invert_yaxis()
ax.set_facecolor('#ebedeb')

fig.suptitle('Viagens por cargo público (2023)')

plt.figtext(0.75, 0.89, 'Fonte: Portal da Transparência', fontsize=8)

plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.yticks(fontsize=8)
plt.xlabel('Número de viagens')
plt.ylabel('Cargo')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(16,6))
ax.barh(df_final['Cargo'], df_final['despesa_media'], color='#1f5c23')
ax.invert_yaxis()
ax.set_facecolor('#ebedeb')

fig.suptitle('Despesas média por cargo público (2023)')

plt.figtext(0.75, 0.89, 'Fonte: Portal da Transparência', fontsize=8)

plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.yticks(fontsize=8)
plt.xlabel('Despesas média')
plt.ylabel('Cargo')
plt.show()
```
